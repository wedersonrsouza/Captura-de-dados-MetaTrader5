```
class Campo:
    def __init__(self, inicio, final):
        self.inicio = inicio
        self.final = final

    def __set_name__(self, owner, nome):
        self.nome = nome

    def __get__(self, instance, owner):
        if not instance:
            return self
        return instance.dados_brutos[self.inicio: self.final]

class Base:
    def __init__(self, dados):
        self.dados_brutos = dados

    def __repr__(self):
        campos = []
        for name, obj in self.__class__.__dict__.items():
            if isinstance(obj, Campo):
                campos.append((name, getattr(self, name)))
        return "\n".join(f"{campo}:{conteudo}" for campo, conteudo in campos)

class HISTORICO(Base):
    tipreg = Campo(0, 2)
    datapregao = Campo(2, 10)
    codbdi = Campo(10, 12)
    codneg = Campo(12,24)
    tpmerc = Campo(24,27)
    nomres = Campo(27,39)
    especi = Campo(39,49)
    prazot = Campo(49,52)
    modref = Campo(52,56)
    preabe = Campo(56,69)
    premax = Campo(69,82)
    premin = Campo(82,95)
    premed = Campo(95,108)
    preult = Campo(108,121)
    preofc = Campo(121,134)
    preofv = Campo(134,147)
    totneg = Campo(147,152)
    quatot = Campo(152,170)
    voltot = Campo(170,188)
    preexe = Campo(188,201)
    indopc = Campo(201,202)
    datven = Campo(202,210)
    fatcot = Campo(210,217)
    ptoexe = Campo(217,230)
    codisi = Campo(230,242)
    dismes = Campo(242,245)
```

```
 def abre_arquivo(nome_arquivo):
   arquivo="hist_acoes/"+nome_arquivo
   print(arquivo)

   conteudo = open(arquivo, 'r', encoding='latin-1').readlines()
   row = conteudo
   return row

#row = abre_arquivo('COTAHIST_A2010.TXT')

#b = HISTORICO(row[2])
#t = b.preabe.format([2])
#print (b.preabe)
#print(float(b.preabe) / 100)
```

```
import pandas as pd
```

```
dados = []

for ano in range(2001,2022):
  
  arquivo="COTAHIST_A{}.TXT".format(ano)
  row = abre_arquivo(arquivo)

  for i in range(1,len(row)-1):
    fetch = HISTORICO(row[i])

    print ("Captando a linha ", i, 'do ano ', ano)
    dados += [{ 
        'symbol': fetch.codneg.strip(),
        'Data': fetch.datapregao.strip(),
        'open': float(fetch.preabe) / 100,
        'high': float(fetch.premax) / 100,
        'low': float(fetch.premin) / 100, 
        'close': float(fetch.preult) / 100,
        'volume': int(fetch.voltot)
    }]

df_dados = pd.DataFrame(dados)
df_dados.head()
```

```
df_dados.to_csv("historicocompleto.csv", sep=';')
```
